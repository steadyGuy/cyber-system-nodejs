```python
### Import libs and setup global settings
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import matplotlib.axes as ax
import numpy as np
import seaborn
# import warnings
from pymongo import MongoClient

plt.rcParams["figure.figsize"] = (22, 10)
pd.set_option('display.max_rows', None)
# warnings.filterwarnings("ignore")
```

```python
### Setup collection from MongoDB

def _connect_mongo(host, port, username, password, db):
    """ A util for making a connection to mongo """

    if username and password:
        mongo_uri = 'mongodb://%s:%s@%s:%s' % (username, password, host, port)
        conn = MongoClient(mongo_uri)
    else:
        conn = MongoClient(host, port)


    return conn[db]


def read_mongo(db, collection, query={}, host='localhost', port=27017, username=None, password=None, no_id=True):
    """ Read from Mongo and Store into DataFrame """

    # Connect to MongoDB
    db = _connect_mongo(host=host, port=port, username=username, password=password, db=db)

    # Make a query to the specific DB and Collection
    cursor = db[collection].find(query)

    # Expand the cursor and construct the DataFrame
    df =  pd.DataFrame(list(cursor))

    # Delete the _id
    if no_id:
        del df['_id']

    return df

df = read_mongo(username='sammy', password='your_password', db='sharkinfo', collection="metricitems")
df.drop(columns=['__v'], inplace = True)
df['windDirection'] = df['windDirection'].astype('int')
```

```python
### Select date/time from and date/time to ###
import ipywidgets as widgets

from_d = widgets.DatePicker(
    description='From',
    disabled=False
)

from_t = widgets.TimePicker(
    description='From',
    disabled=False
)

to_d = widgets.DatePicker(
    description='To',
    disabled=False
)

to_t = widgets.TimePicker(
    description='To',
    disabled=False
)

box = widgets.Box(children=[from_d, from_t, to_d, to_t])
box
```

```python
### Assigning date to a variables

date_from = datetime.datetime.combine(from_d.value, 
                          from_t.value)

date_to = datetime.datetime.combine(to_d.value, 
                          to_t.value)

new_df = (df.date > date_from) & (df.date < date_to)
date_df = df[new_df]
date_df
```

```python
plt.grid()
plt.xlabel('Date')
plt.ylabel('T - °C')


x = date_df.date
y = date_df.temperature
# plt.show()

x_ticks = [15340., 15400., 15461., 15522., 15584., 15645., 15706.]

#specify x-axis labels
x_labels = [15340., 15400., 15461., 15522., 15584., 15645., 15706.]

plt.plot(x, y, linewidth = 2, label='Temperature conditions of the region')
plt.legend()
# plt.xticks(np.arange(min(x), max(x)+1, 1.0))
# locs, labels = plt.xticks()
# print(locs, labels)
plt.show()
```

```python
temperature_elements = date_df['temperature'].value_counts().sample(frac=1).apply(lambda i: i * 0.5)

def valuelabel(x, y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center',
                 bbox = dict(facecolor = 'lightblue', alpha = 0.8))

temperatures = temperature_elements.index
temperatures_count = temperature_elements.values

fig, ax = plt.subplots(figsize = (22,10))

idx = np.asarray([i for i in range(len(temperatures))])
ax.bar(idx, [i for i in temperatures_count])

ax.set_xticks(idx)
ax.set_xticklabels(temperatures)
ax.legend(['Duration of temperature regimes'])
ax.set_xlabel('T - °C')
ax.set_ylabel('t, год')
valuelabel(temperatures, temperatures_count)  

fig.tight_layout()

plt.show()
```

```python
def speed_labels(bins, units):   
    labels = []
    for left, right in zip(bins[:-1], bins[1:]):
        if left == bins[0]:
            labels.append('calm'.format(right))
        elif np.isinf(right):
            labels.append('>{} {}'.format(left, units))
        else:
            labels.append('{} - {} {}'.format(left, right, units))

    return list(labels)

total_count = date_df.shape[0]
calm_count = date_df.query("windSpeed <= 0").shape[0]
max_speed = date_df['windSpeed'].max()
dir_bins = np.arange(0, 365, 45)
spd_bins = [-np.inf, 0, max_speed * 0.25, max_speed * 0.5, max_speed * 0.75, max_speed]
dir_labels = [45, 90, 135, 180, 225, 270, 315, 360]
spd_labels = speed_labels(spd_bins, units='')

def _convert_dir(directions, N=None):
    if N is None:
        N = directions.shape[0]
    barDir = directions * np.pi/180. - np.pi/N
    barWidth = 2 * np.pi / N
    return barDir, barWidth

def wind_rose(rosedata, wind_dirs, palette=None):
    if palette is None:
        palette = seaborn.color_palette('inferno', n_colors=rosedata.shape[1])

    bar_dir, bar_width = _convert_dir(wind_dirs)
    fig, ax = plt.subplots(figsize=(22, 10), subplot_kw=dict(polar=True))
    ax.set_theta_direction('clockwise')
    ax.set_theta_zero_location('N')

    for n, (c1, c2) in enumerate(zip(rosedata.columns[:-1], rosedata.columns[1:])):
        if n == 0:
            # first column only
            ax.bar(bar_dir, rosedata[c1].values, 
                   width=bar_width,
                   color=palette[0],
                   edgecolor='none',
                   label=c1,
                   linewidth=0)
        # all other columns
        ax.bar(bar_dir, rosedata[c2].values, 
               width=bar_width, 
               bottom=rosedata.cumsum(axis='columns')[c1].values,
               color=palette[n+1],
               edgecolor='none',
               label=c2,
               linewidth=0)

    leg = ax.legend(loc=(0.75, 0.95), ncol=2)
    xtl = ax.set_xticklabels(['N', 'NE', 'E', 'SE', 'S', 'SW', 'W', 'NW'])
    
    return fig
dir_counts = list(date_df.assign(WindSpd_bins=lambda df: pd.cut(df['windSpeed'], bins=spd_bins, labels=spd_labels, right=True)).groupby('WindSpd_bins').size())

dir_counts_percent = [str(round(i / sum(dir_counts) * 100, 2)) for i in dir_counts]
spd_labels = [''.join([spd_labels[i], str('%' + dir_counts_percent[i])]) for i in range(len(dir_counts_percent))]
print('dir_counts_percent', spd_labels)

rose = (date_df.assign(WindSpd_bins=lambda df: pd.cut(df['windSpeed'], bins=spd_bins, labels=spd_labels, right=True))
                 .replace({'windDirection': {360: 0}})
                 .groupby(by=['WindSpd_bins', 'windDirection'])
                 .size()
                 .unstack(level='WindSpd_bins')
                 .fillna(0)
                 .assign(calm=lambda df: calm_count / df.shape[0])
                 .sort_index(axis=1)
                 .applymap(lambda x: x / total_count * 100)
          )
rose.drop(columns=[spd_labels[0]], inplace = True)
rose = rose.rename({'calm': spd_labels[0]}, axis='columns')
# pd.qcut(date_df['windSpeed'], q=4)
list(df['windDirection'].unique())
fig = wind_rose(rose, np.arange(0, 360, 45))
rose
```

```python
wind_elements = date_df['windSpeed'].value_counts().apply(lambda i: i * 0.5)

wind_speed = wind_elements.index
wind_hours = wind_elements.values

fig, ax = plt.subplots(figsize = (22,10))

idx = np.asarray([i for i in range(len(wind_speed))])
ax.bar(idx, [i for i in wind_hours])

ax.set_xticks(idx)
ax.set_xticklabels(wind_speed)
ax.legend(['Duration of temperature regimes'])
ax.set_xlabel('м / с')
ax.set_ylabel('t, год')
valuelabel(wind_speed, wind_hours)  

fig.tight_layout()
plt.grid(axis = 'y')

plt.show()
```
